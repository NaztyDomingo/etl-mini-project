```python
smhi_df = pd.read_csv('smhi_weather_data.csv')
met_df = pd.read_csv('met_weather_data.csv')
```

```python
smhi_df['source'] = 'SMHI'
met_df['source'] = 'MET'
```

```python
smhi_df['location'] = smhi_df['location'].replace({
    '57.7_11.9': 'Gothenburg',
    '55.6_13.0': 'Malmo',
    '59.9_10.7': 'Oslo',
    '59.3_18.0': 'Stockholm',
    '60.1_24.9': 'Helsinki'
})

met_df['location'] = met_df['location'].replace({
    '57.7_11.9': 'Gothenburg',
    '55.6_13.0': 'Malmo',
    '59.9_10.7': 'Oslo',
    '59.3_18.0': 'Stockholm',
    '60.1_24.9': 'Helsinki'
})
```

```python
smhi_df['cloud_coverage'] = (smhi_df['cloud_coverage'] / 8) * 100
```

```python
smhi_df['time'] = pd.to_datetime(smhi_df['time'])
smhi_df['day_of_week'] = pd.to_datetime(smhi_df['time']).dt.day_name()
smhi_df['time_of_day'] = pd.to_datetime(smhi_df['time']).dt.hour
smhi_df['time'] = smhi_df['time'].dt.strftime('%Y-%m-%d')
```

```python
met_df['time'] = pd.to_datetime(met_df['time'])
met_df['day_of_week'] = pd.to_datetime(met_df['time']).dt.day_name()
met_df['time_of_day'] = pd.to_datetime(met_df['time']).dt.hour
met_df['time'] = met_df['time'].dt.strftime('%Y-%m-%d')
```

```python
smhi_df.sample(30)
```

```python
met_df.sample(30)
```

```python
met_df.pivot_table(index=['location', 'time', 'time_of_day'], values=['temperature', 'cloud_coverage', 'relative_humidity', 'wind_speed'], aggfunc='mean').round(2)
```
